# A Simple Neural Network from Scratch

a_simple_nn.png

```python
import numpy as np
```

## Initialize Weights and Bias

```python
W1 = [[0.1, 0],
      [0.3, 0.1],
      [0.1, 0.1],
      [0, 0.9]]

b1 = np.full((4, 1), -1)

W2 = [[0, 0.33, 0.2, 0.3],
      [0.33, 0, 0, 0],
      [0, 0.5, 0, 0.33],
      [0.33, 0, 0.3, 0]]

b2 = np.full((4, 1), 1)

# Assuming weights for last layer to be 1
W3 = np.ones((1, 4))
b3 = np.full((1, 1), 2)
```

## Define Activation Functions

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))

def relu(z):
  return np.maximum(0, z)

def tanh(z):
  return np.tanh(z)

activations = {
    "sigmoid": sigmoid,
    "tanh": tanh,
    "relu": relu
}
```

### Define Forward Pass

```python
def forward(X, type="sigmoid"):
  # Layer 1
  Z1 = np.dot(W1, X) + b1
  Y1 = activations[type](Z1)

  # Layer 2
  Z2 = np.dot(W2, Y1) + b2
  Y2 = activations[type](Z2)

  # Output Layer
  Z3 = np.dot(W3, Y2) + b3
  Y3 = activations[type](Z3)

  return Y3
```

## Part 1:

x1 = Last two digits of student ID

x2 = Date of birth

Activation Function = Sigmoid on all nodes

### Output

```python
X1 = [[2], [10]]

Y = forward(X1, type="sigmoid")
print(Y)
```

## Part 2:


x1 = Good

x2 = Day

Activation Function = Tanh on all nodes

### Output

```python
# One hot encoding for input X1 = Good, X2 = Day
X2 = [[1, 0], [0, 1]]

Y2 = forward(X2, type="tanh")
print(Y2)
```

## Part 3:


x1 = 23

x2 = 25

Activation Function = ReLU on all nodes

### Output

```python
X3 = [[23], [25]]

Y3 = forward(X3, type="relu")
print(Y3)
```
