## Reflex Agent for Cleaning



```python
class ReflexCleaningAgent:
    def __init__(self):
        self.water_used = 0
        self.co2_emission = 0

    def perceive(self, waste_present):
        return waste_present

    def act(self, waste_present):
        if waste_present:
            print("Waste detected: Cleaning")
            self.water_used += 1   # minimal water usage
            self.co2_emission += 0.5  # small emission
        else:
            print("No waste: Staying idle (eco mode)")

    def status(self):
        print(f"Water used: {self.water_used}")
        print(f"CO2 emitted: {self.co2_emission}")


```

```python
# Test run
agent = ReflexCleaningAgent()

environment = [True, False, True, False]

for state in environment:
    agent.act(agent.perceive(state))

agent.status()
```

## Utility-based Agent for Earning Airline Points

utility = profit - (CO2 penalty)

```python
class AirlineUtilityAgent:
    def __init__(self):
        pass

    def utility(self, profit, co2):
        return profit - (2 * co2)  # penalize emissions

    def choose_flight(self, options):
        best_option = None
        best_utility = float('-inf')

        for option in options:
            u = self.utility(option['profit'], option['co2'])
            print(f"Option {option['name']}: Utility: {u}")

            if u > best_utility:
                best_utility = u
                best_option = option

        return best_option


```

```python
# Example flight options
flights = [
    {"name": "Flight A", "profit": 100, "co2": 40},
    {"name": "Flight B", "profit": 80, "co2": 20},
    {"name": "Flight C", "profit": 120, "co2": 70},
]

agent = AirlineUtilityAgent()
best = agent.choose_flight(flights)

print(f"\nBest choice: {best['name']}")
```

## Goal-based agent in In-Drive taxi to provide secure commute

We choose routes based on:

* Safety
* Distance (less fuel -> less CO₂)



```python
class InDriveTaxiGoalAgent:
    def __init__(self, goal="safe_commute"):
        self.goal = goal

    def evaluate_route(self, route):
        score = 0

        # prioritize safety
        if route["safe"]:
            score += 50

        # shorter distance → better
        score += (100 - route["distance"])

        return score

    def choose_route(self, routes):
        best_route = None
        best_score = float('-inf')

        for route in routes:
            score = self.evaluate_route(route)
            print(f"Route {route['name']}: Score: {score}")

            if score > best_score:
                best_score = score
                best_route = route

        return best_route


```

```python
# Example routes
routes = [
    {"name": "Route A", "distance": 30, "safe": True},
    {"name": "Route B", "distance": 20, "safe": False},
    {"name": "Route C", "distance": 25, "safe": True},
]

agent = InDriveTaxiGoalAgent()
best = agent.choose_route(routes)

print(f"\nBest route: {best['name']}")
```
